```python
import pandas as pd
import matplotlib.pyplot as plt
path_csv = 'https://raw.githubusercontent.com/DavideScassola/data_analysis_laboratory/main/notebooks/dati/reviews.csv'
reviews = pd.read_csv(path_csv)
```

```python
# 1
# Trova il videogioco con il punteggio più basso
reviews[reviews['score'] == reviews['score'].min()]
```

```python
# 2
# Trova il videogioco di genere 'Shooter' con il punteggio più alto
s_reviews = reviews[reviews['genre'] == 'Shooter']
s_reviews[(s_reviews['score'] == s_reviews['score'].max())]
```

```python
# 3
# Trova il videogioco più vecchio
reviews[reviews['release_year'] == reviews['release_year'].min()]
```

```python
# 4
# Quanti videogiochi per PC hanno avuto un voto superiore a 9 nel 2005? Consiglio: applica la funzione `len()` per ottenere il numero di righe di un DataFrame
len(reviews[(reviews['platform'] == 'PC') & (reviews['score'] > 9) & (reviews['release_year'] == 2005) ])
```

```python
reviews[(reviews['platform'] == 'PC') & (reviews['score'] > 9) & (reviews['release_year'] == 2005) ]
```

```python
# 5
# Trova il videogioco più nuovo
# Consiglio: puoi ordinare le righe di un DataFrame in base al valore di una colonna (o più) con il metodo sort_values.

#reviews.sort_values(['release_year', 'release_month', 'release_month'], ascending=False).iloc[0]

last_year_reviews = reviews[reviews['release_year'] == reviews['release_year'].max()]
last_month_reviews = last_year_reviews[last_year_reviews['release_month'] == last_year_reviews['release_month'].max()]
last_day_reviews = last_month_reviews[last_month_reviews['release_day'] == last_month_reviews['release_day'].max()]
last_day_reviews
```

```python
# 6
# Come hai notato il dataset ha sia una colonna score che una colonna score_phrase (il voto).
# Riesci a trovare a quale score_phrase corrisponde ogni punteggio?
# In altre parole, quali sono le soglie di punteggio per passare da un voto al successivo?

# Consiglio: puoi usare il metodo .unique() per trovare i valori unici di una serie
reviews['score_phrase'].unique()

for score in reviews['score_phrase'].unique():
    minimo = reviews[reviews['score_phrase'] == score]['score'].min()
    massimo = reviews[reviews['score_phrase'] == score]['score'].max()
    print(f"{score}: da {minimo} a {massimo}")
```

```python
# 7
# Carica il dataset dal .csv che si trova al seguente path
path = 'sample_data/california_housing_train.csv'

df = pd.read_csv(path)
df
```

```python
# 8
# Mostra un istogramma della variabile 'median_house_value'
_ = plt.hist(df['median_house_value'], bins=30)
```

```python
# 9
# Mostra un grafico (scatterplot) della variabile 'population' (asse x) rispetto alla variabile 'total_rooms' (asse y)
# Consiglio: puoi usare il metodo 'scatter' di matplotlib: funziona in modo simile a 'plot' ma non unisce i punti con una linea di default.
_ = plt.scatter(df['population'], df['total_rooms'], alpha=0.4)
```

```python
# 10
# Quanti abitanti ci sono per ogni casa?
df['population'].sum() / df['households'].sum()
```

```python
# 11
# Questo dataset contiene le coordinate geografiche di ogni quartiere, riesci a mostrare in un grafico la posizione di ogni quartiere? (come se fosse una mappa)
_ = plt.scatter(df['longitude'], df['latitude'], alpha=0.4, color='orange')
```

```python
# 12
# Mostra come prima la posizione di ogni quartiere, ma questa volta i quartieri con meno di 5000 abitanti devono essere rapresentati con un colore diverso rispetto ai quartieri con più di 5000 abitanti
# Consiglio: se chiami più di una volta una funzione per fare grafici di matplotlib (come ply.plot o plt.scatter), i due grafici verranno sovrapposti

pop = 5000
df_big = df[df['population'] > pop]
df_small = df[df['population'] <= pop]

_ = plt.scatter(df_small['longitude'], df_small['latitude'], alpha=0.5, color='blue', label=f'popolazione <= {pop}')
_ = plt.scatter(df_big['longitude'], df_big['latitude'], alpha=0.5, color='orange', label=f'popolazione > {pop}')
_ = plt.legend()
```
